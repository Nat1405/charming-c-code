```python
# Based on https://docs.opencv.org/3.1.0/d4/d13/tutorial_py_filtering.html

import cv2
import numpy as np
from IPython.display import clear_output, Image, display
from io import BytesIO
import PIL.Image

def showarray(a, fmt='jpeg'):
    a = np.uint8(np.clip(a, 0, 255))
    f = BytesIO()
    PIL.Image.fromarray(a).save(f, fmt)
    display(Image(data=f.getvalue()))
    
```

```python
# Do an averaging

img = cv2.imread("newspaper2.jpg")
showarray(img)
#kernel = np.ones((5,5),np.float32)/25
#print(kernel)
blur = cv2.blur(img,(5,5))
showarray(blur)
```

```python
# Do a Gaussian blur

img = cv2.imread("newspaper2.jpg")
showarray(img)
kernel = cv2.getGaussianKernel(5,5)
print(kernel)
blur = cv2.GaussianBlur(img,(5,5),0)
showarray(blur)
```

```python
# Do a median blur

img = cv2.imread("newspaper2.jpg")
showarray(img)
median = cv2.medianBlur(img,5)
showarray(median)
```

```python
# Do a bilateral blur

img = cv2.imread("newspaper2.jpg")
showarray(img)
blur = cv2.bilateralFilter(img,9,150,150)
showarray(blur)
```

```python
# Downsample then bilateral blur
img = cv2.imread("newspaper2.jpg")
showarray(img)
downsampled = cv2.pyrDown(img, dstsize=(img.shape[0]//2, img.shape[1]//2))
downsampled = cv2.pyrUp(downsampled, dstsize=(downsampled.shape[0]*2, downsampled.shape[1]*2))
showarray(downsampled)
# Do the bilateral blur
blur = cv2.bilateralFilter(downsampled,9,150,150)
showarray(blur)
```
